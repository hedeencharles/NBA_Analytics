```python
import pandas as pd
import glob
import os
os.chdir("../Data/Data_Raw")
```

```python
# Creating a list of file names in the directory
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
# Merging all CSVs together
master_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
master_csv
```

```python
# Finding the number of unique players in our combined dataset
master_csv['Player'].nunique()
```

```python
master_csv.sort_values('Player').head(20)
```

```python
# Dropping rows that contain 'TOT' so I can combine rows for accurate totals
master_csv = master_csv[master_csv.Tm != 'TOT']

master_csv
```

```python
# Slimming down the data to get what I want for my heatmap DB
win_share_df = master_csv[['Player','WS']]

win_share_df = win_share_df.groupby(['Player']).sum().reset_index()

win_share_df.sort_values('WS', ascending=False).head(30)
```
